## Factor Mimicking Portfolios

### Introduction
Factor mimicking portfolios are constructed to capture systematic risk factors that explain the returns of a set of assets. These portfolios serve as proxies for theoretical factors in asset pricing models, such as the Fama-French three-factor model, which includes market, size, and value factors.

### Objective
Given the loadings matrix \( B \) and a set of asset returns \( R \), our goal is to estimate the factor returns \( F \) and construct the portfolio weights \( V \) that mimic these factors.

### Estimation Problem
The estimation problem can be formulated as a weighted least squares problem:

\[
\min (R - B F)^\top W (R - B F) \quad \text{subject to} \quad F \in \mathbb{R}^m
\]

where:
- \( R \) is the vector of asset returns.
- \( B \) is the factor loadings matrix.
- \( F \) is the vector of factor returns to be estimated.
- \( W \) is a diagonal and positive definite weighting matrix.

### Weighting Matrix
The matrix \( W \) is typically set to the inverse of the idiosyncratic covariance matrix \( \Omega \). However, since \( \Omega \) is not known in advance, a common practical approach is to use the weights:

\[
W_{ii} = \left( \text{market cap}_i \right)^{1/2}
\]

where the market cap of the stock is averaged over a number of days (e.g., 21 trading days).

### Solution

#### Estimated Factor Returns
The estimated factor returns \( \hat{F} \) can be obtained using the formula:

\[
\hat{F} = (B^\top W B)^{-1} B^\top W R
\]

#### Residuals
The residuals \( \hat{\epsilon} \) can be calculated as:

\[
\hat{\epsilon} = (I - B(B^\top W B)^{-1} B^\top W) R
\]

#### Portfolio Weights
The matrix \( V^\top \), whose rows represent the portfolio weights for the factor-mimicking portfolios, is given by:

\[
V^\top = (B^\top W B)^{-1} B^\top W
\]

Each row of the matrix \( V^\top \) represents a vector of portfolio weights that correspond to a specific factor. This portfolio has a return \( \hat{F} \) and is called the factor-mimicking portfolio for factor \( i \). We denote \( V \) as the matrix whose \( i \)-th column is the vector \( v_i \).

### Interpretation
The matrix \( V^\top \) can be interpreted as follows:
- Each row \( V_i^\top \) represents the weights of a minimum-variance portfolio that has unit exposure to factor \( i \) and zero exposure to any other factor.
- The estimated factor returns \( \hat{F} \) are the scalar product of the \( i \)-th row of the matrix \( V^\top \) and the return vector \( R \).

### Python Implementation

Here's a Python implementation using NumPy and pandas:

```python
import numpy as np
import pandas as pd

class FactorMimickingPortfolios:
    def __init__(self, B, W):
        self.B = B
        self.W = W

    def calculate_factor_returns(self, R):
        B_T = self.B.T
        factor_returns = np.linalg.inv(B_T @ self.W @ self.B) @ B_T @ self.W @ R
        return factor_returns

    def calculate_residuals(self, R):
        B_T = self.B.T
        P = B_T @ self.W @ self.B
        H = self.B @ np.linalg.inv(P) @ B_T @ self.W
        residuals = (np.eye(self.B.shape[0]) - H) @ R
        return residuals

    def calculate_portfolio_weights(self):
        B_T = self.B.T
        W_inv = self.W
        portfolio_weights = np.linalg.inv(B_T @ W_inv @ self.B) @ B_T @ W_inv
        return portfolio_weights.T

# Example data
B = np.array([
    [1.0, 0.5],
    [0.8, 0.3],
    [0.3, 0.7]
])

market_caps = np.array([100, 200, 150])
W = np.diag(np.sqrt(market_caps))

R = np.array([0.05, 0.07, 0.04])

fmp = FactorMimickingPortfolios(B, W)

# Calculate factor returns
F = fmp.calculate_factor_returns(R)
print("Factor Returns (F):")
print(F)

# Calculate residuals
residuals = fmp.calculate_residuals(R)
print("\nResiduals (Epsilon):")
print(residuals)

# Calculate portfolio weights
V = fmp.calculate_portfolio_weights()
print("\nPortfolio Weights (V):")
print(V)
```

$$
\frac{\partial}{\partial F} (R^\top W R) = 0

$$

\frac{\partial}{\partial F} (F^\top B^\top W B F) = 2 B^\top W B F

$$
\frac{\partial L(F)}{\partial F} = - 2 B^\top W R + 2 B^\top W B F
